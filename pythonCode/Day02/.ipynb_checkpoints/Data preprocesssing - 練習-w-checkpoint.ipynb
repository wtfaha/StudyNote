### 04.資料對應\取代\分箱

```python
#map{key01,key02}

df = pd.DataFrame({'gender':['male','male','female','male','female','female']})
df
```

```python
gender_to_boolean = {'female':0,'male':1} #dict
df['code'] = df['gender'].map(gender_to_boolean)
df
```

```python
#取代

df = pd.DataFrame({'col1':['c01','c02','c03','c04','c05'],
                   'col2':[65,'NULL','-','-',78],
                   'col3':[321,34,'NULL','-',34]})
df
```

```python
df['col2-new'] = df['col2'].replace('-',0)
```

```python
df
```

```python
df.replace('NULL',0)
```

```python
df.replace(['-','NULL'],0)
```

```python
df = df.replace({'-':0,'NULL':-1})  #dictionary
df
```

```python
# 分箱 bins
```

```python
df = pd.DataFrame({'id':['John','Mary','Tom','Nick','Alice'],
                   'score':[90,59,68,77,60]})
df
```

```python
# 切割
bins = [0,60,70,80,90,100]
pd.cut(df['score'],bins)
```

```python
# 邊界值
pd.cut(df['score'],bins, right=False)
```

```python
# label 截距
bins = [0,60,70,80,90,100]
labels = ["F","D","C","B","A"] # 截距
pd.cut(df['score'],bins, right=False, labels=labels)
```

```python
df['label'] = pd.cut(df['score'],bins, right=False, labels=labels)
df
```

### Pandas 合併dataframe

```python
#concat
import pandas as pd
import numpy as np

#定義資料集
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2, columns=['a','b','c','d'])
```

```python
#concat縱向合併
res = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
res
```

```python
# merge
import pandas as pd

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K1', 'K2', 'K3', 'K4'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# merge 會少掉很多資料
res = pd.merge(left, right, on="key")
res
```

```python
# merge
import pandas as pd

#定義資料集並打印出
left = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],  'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],   'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
#依key1與key2 進行合併，並印出四種結果['left', 'right', 'outer', 'inner']
res = pd.merge(left, right, on=["key1","key2"], how="left")
res
```
